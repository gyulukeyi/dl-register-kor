````python
import pickle
import random
````

````python
import openai

with open('.creds') as f:
    openai.api_key = f.read()
````

````python
with open('inputs_test_50.pkl', 'rb') as f:
    testsets = pickle.load(f)
````

````python
import re 

answers = list()

make_fiveshot = lambda testset: ''.join(['    - '+t0+' (class 0)\n'+'    - '+t1+' (class1)\n' for t0, t1 in zip(testset['class_0'][:5], testset['class_1'][:5])])
make_texts = lambda testset0, testset1: '\n'.join(
    random.sample(
    [f"<0>    - {t0}" for t0 in testset0] +
      [f"<1>    - {t1}" for t1 in testset1],
    len(testset0)+len(testset1))
)
extract_answersheet = lambda t: re.sub(r"^<\d>", '', t, flags=re.MULTILINE) if (answers.extend([
    int(re.search(r'\d', u).group(0)) for u in re.findall(r"^<\d>", t, flags=re.MULTILINE)
])) else re.sub(r"^<\d>", '', t, flags=re.MULTILINE)
make_inputs = lambda testset: [
    f"""
Classify each texts below in accordance to the samples with 0 or 1:

Samples:
{make_fiveshot(testset)}

Texts:
{extract_answersheet(
    make_texts(testset['class_0'][10+10*i:10+10*(i+1)], testset['class_1'][10+10*i:10+10*(i+1)])
)}
    """ for i in range(int(50/10)-1)]
````

````python
print(make_inputs(testsets['isMaleTest50'])[2])
````

````python
import time
from tqdm import tqdm

result = dict() # {feature-run(str): responseObj(list)}

for testset in tqdm(testsets.keys(), desc='Iterating testsets...',
                    position=0):
    # testset: feature name
    answers = list()
    # list of Y to be filled by make_inputs 
    # @ISSUE0001 `answers` being initiated here make each iteration of `make_inputs`,
    #   which will shuffle inputs, accumulate all labels to `answers` variable
    #   => @FIX0001 will divide each `answers` into three respective `answers`
    runs = 1
    while runs <= 3:
    # iter three times to normalize GPT answer 
        response = list()   
        # storage for GPT response
        for prompt in tqdm(make_inputs(testsets[testset]),
                        desc=f"Iteraing: {testset}...",
                        position=1):
            # @ISSUE0002 each iteration of `make_inputs` will shuffle items per run 
            #     thus making normalization difficult 
            #   
            while True:    
                try:
                    response.append(openai.ChatCompletion.create(
                        model="gpt-3.5-turbo",
                        messages=[
                            {
                                'role': 'system',
                                'content': 'You are classifying each of the texts into two given classes.'
                            },
                            {
                                'role': 'user',
                                'content': prompt,
                            }
                        ],
                        temperature=1,
                        max_tokens=501,
                        top_p=1,
                        frequency_penalty=0,
                        presence_penalty=0
                        )
                    )
                    break
                except KeyboardInterrupt:
                    print("Halting upon keyboard interrupt...")
                    break
                except BaseException as e:
                    print("Received error:")
                    print(e)
                    print("Continue after a while.")
                    print("blame OpenAI")
                    time.sleep(10)
                
        result[f"{testset}-{runs}"] = response
        result[f"{testset}-label"] = answers 
        runs += 1
````

````python
result.keys()
````

````python
# @FIX0001 -- adjust accumulated labels (@ISSUE0001)
adjust_out = lambda feature: {
    f'is{feature}Test50-1-label': result[f"is{feature}Test50-label"][80*0:80*1], 
    f'is{feature}Test50-2-label': result[f"is{feature}Test50-label"][80*1:80*2], 
    f'is{feature}Test50-3-label': result[f"is{feature}Test50-label"][80*2:80*3], 
}
result.update(adjust_out('Male'))
result.update(adjust_out('Family'))
result.update(adjust_out('Youth'))
result.keys()
````

````python
make_out_preds = lambda key: [x 
                              for choices in result[key] 
                              for x in choices['choices'][0]['message']['content'].split('\n') 
                              ]
make_out_labels = lambda key: result[key]

outs = dict()
for key in result.keys():
    outs[key] = make_out_preds(key) if 'label' not in key else make_out_labels(key)
````

````python
assert (True
    and type(outs) == dict
    and all([type(x)==list for x in outs.values()])
    and all([len(x)>=80 for x in outs.values()])
)
````

````python
import json
with open('gpt_outs_toclean.json', 'w') as f:
    json.dump(outs, f, ensure_ascii=False, indent=4)
````

JSON file cleaned with the following expressions:

```bash
s/"Texts:",\n\s*//g
s/"",\n\s*//g
s/".+class ([01]).*"/$1/g
s/".+([01])",/$1,/g
s/^\s*".+?:",\n//g
s/^\s*"\s+",\n//g
s/".+",/null/g
```

Note that the expressions were applied in given order.

Also note that 5 NULLs were made to denote the cases where no class can be determined from GPT generation.

Cleaned file saved as `gpt_out_cleaned.json`

`isYouthTest50` run results are consistent as 79 per run, but a sample run on `make_inputs` confirms that 80 inputs were passed indeed. 

However, since we cannot determine which of the input was ignored as the generation was not consistent in reporting the labeling text. 

A manual inspection with inconsistent generation reveals the following text is ignored:

> - 근데 미쳤나 봐. 그거 그거 끝나자마자 갑자기 목에 담이 온 거예요.

Assumption is that the generation for the line occurred, but censored due to the word '미쳤나'.

Since it is unable to track the case, run iteration again only for the Youth feature, excluding '미쳤나' case.

Also, `isFamilyTest50-3` included 81 predictions. An inspection on the generation results revealed that the API returned a class for 21th text for the third batch, which were not given. Thus, prediction for the 21th text was ignored by manual deletion.

````python
import time
from tqdm import tqdm

LF = '\n' 

make_inputs_filter_crazy = lambda testset: [
    f"""
Classify each texts below in accordance to the samples with 0 or 1:

Samples:
{make_fiveshot(testset)}

Texts:
{extract_answersheet(
    LF.join([t for t in 
     make_texts(testset['class_0'][10+10*i:10+10*(i+1)], testset['class_1'][10+10*i:10+10*(i+1)]).split(LF)
     if '미쳤나' not in t 
    ])
)}
    """ for i in range(int(50/10)-1)]

result_youth = dict() # {feature-run(str): responseObj(list)}

testset = 'isYouthTest50'
runs = 1
while runs <= 3:
# iter three times to normalize GPT answer 
    answers = list()
    response = list()   
    # storage for GPT response
    for prompt in tqdm(make_inputs_filter_crazy(testsets[testset]),
                    desc=f"Iteraing: {testset}...",
                    position=1):
        while True:       
            try:
                response.append(openai.ChatCompletion.create(
                    model="gpt-3.5-turbo",
                    messages=[
                        {
                            'role': 'system',
                            'content': 'You are classifying each of the texts into two given classes.'
                        },
                        {
                            'role': 'user',
                            'content': prompt,
                        }
                    ],
                    temperature=1,
                    max_tokens=501,
                    top_p=1,
                    frequency_penalty=0,
                    presence_penalty=0
                    )
                )
                break
            except KeyboardInterrupt:
                print("Halting upon keyboard interrupt...")
                break
            except BaseException as e:
                print("Received error:")
                print(e)
                print("Continue after a while.")
                print("blame OpenAI")
                time.sleep(10)
            
    result_youth[f"{testset}-{runs}"] =  response
    result_youth[f"{testset}-{runs}-label"] = answers 
    runs += 1
````

````python
make_out_preds_youth = lambda key: [x 
                              for choices in result_youth[key] 
                              for x in choices['choices'][0]['message']['content'].split('\n') 
                              ]
make_out_labels_youth = lambda key: result_youth[key]

outs_youth = dict()
for key in result_youth.keys():
    outs_youth[key] = make_out_preds_youth(key) if 'label' not in key else make_out_labels_youth(key)

pprint([
    (k, len(v)) for k, v in outs_youth.items()
])
````
